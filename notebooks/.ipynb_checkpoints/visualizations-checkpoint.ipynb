```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned/top50_clean.csv")  

df.head()
```

```python
# H1: The most popular songs belong to the Pop genre
#1) Examine the average popularity score by genre_group
```

```python
genre_pop = (
    df.groupby("Genre_Group")["Popularity"]
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8, 5))
genre_pop.plot(kind="bar")
plt.title("Average Popularity by Genre Group")
plt.xlabel("Genre Group")
plt.ylabel("Average Popularity Score")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
#2) Check the genre distribution within the Top 20 songs
```

```python
top20 = df[df["Rank"] <= 20]

top20_counts = (
    top20["Genre_Group"]
    .value_counts()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8, 5))
top20_counts.plot(kind="bar")
plt.title("Genre Group Distribution in Top 20 Songs")
plt.xlabel("Genre Group")
plt.ylabel("Number of Songs in Top 20")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
#H2: Songs with higher danceability are more popular
#1) Categorize Danceability into Low / Medium / High and compare their average popularit
```

```python
def dance_level(x):
    if x < 40:
        return "Low"
    elif x <= 70:
        return "Medium"
    else:
        return "High"

df["Dance_Level"] = df["Danceability"].apply(dance_level)

dance_pop = (
    df.groupby("Dance_Level")["Popularity"]
      .mean()
      .reindex(["Low", "Medium", "High"])
)

plt.figure(figsize=(6, 4))
dance_pop.plot(kind="bar")
plt.title("Average Popularity by Danceability Level")
plt.xlabel("Danceability Level")
plt.ylabel("Average Popularity Score")
plt.tight_layout()
plt.show()
```

```python
#2) Examine whether the songs with the highest danceability also have high popularity scores
```
